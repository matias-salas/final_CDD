```python
df_comuna = pd.read_csv('Dataframe_sin_pct.csv')
```

```python
df_comuna = df_comuna.rename(columns = {'Nombre_de_departamentos_comuna':'nombre_comuna',
                                  'Codigo_de_departamentos_comuna':'codigo_comuna',
                                  'Nombre_de_provincia':'nombre_provincia',
                                  'Provincia':'distrito',
                                  'Departamento':'departamento'})
```

```python
df_comuna = df_comuna.groupby(['distrito','nombre_provincia','departamento','nombre_comuna','codigo_comuna'],as_index=False).sum()
df_comuna.head()
```

### Porcentajes necesarios segun esta img



```python
pd.options.display.max_info_columns
df_comuna.info(verbose=True)
```

```python
df_comuna_remake = df_comuna.copy()
```

```python
df_comuna_remake['hogares_calidad_de_construccion'] = df_comuna_remake['Hogares_con_calidad_de_la_vivienda_precaria'] / df_comuna_remake['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_calidad_de_construccion'] = df_comuna_remake['hogares_calidad_de_construccion'].round(2)
```

```python
df_comuna_remake['hogares_jefes_universitario'] = df_comuna_remake['Hogares_con_jefes_con_universitario_completa'] / df_comuna_remake['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_jefes_universitario'] = df_comuna_remake['hogares_jefes_universitario'].round(2)
```

```python
df_comuna_remake['hogares_personal_con_cama'] = df_comuna_remake['Hogares_con_personas_de_servicio_domestico_sin_retiro_con_cama'] / df_comuna['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_personal_con_cama'] = df_comuna_remake['hogares_personal_con_cama'].round(2)
```

```python
df_comuna_remake['hogares_piso_precario'] = df_comuna_remake['Hogares_sin_piso_ok'] / df_comuna['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_piso_precario'] = df_comuna_remake['hogares_piso_precario'].round(2)
```

```python
df_comuna_remake['personas_utilizan_compu'] = df_comuna_remake['Poblacion_de_18_a_29_anios_que_utiliza_computadora'] / df_comuna['Poblacion_de_18_a_29_anios'] * 100
df_comuna_remake['personas_utilizan_compu'] = df_comuna_remake['personas_utilizan_compu'].round(2)
```

```python
df_comuna_remake['hogares_servicios'] = df_comuna_remake['Hogares_con_desague_a_hoyo_o_pozo_ciego_sin_camara'] / df_comuna['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_servicios'] = df_comuna_remake['hogares_servicios'].round(2)
```

```python
df_comuna_remake['hogares_jefe_con_secundaria'] = df_comuna_remake['Hogares_con_jefes_con_secundaria_completa'] / df_comuna['Cantidad_total_de_hogares'] * 100
df_comuna_remake['hogares_jefe_con_secundaria'] = df_comuna_remake['hogares_jefe_con_secundaria'].round(2)
```

```python
df_comuna_remake['personas_jovenes_asistencia_sin_secundario'] = df_comuna_remake['Personas_14a25_sin_secundario_completoomas_y_quenoestudian'] / df_comuna['Poblacion_de_14_a_25_anios'] * 100
df_comuna_remake['personas_jovenes_asistencia_sin_secundario'] = df_comuna_remake['personas_jovenes_asistencia_sin_secundario'].round(2)
```

```python
df_comuna_remake.to_csv('Dataframe_con_pct_comunas.csv',index=None)
```

```python
df_comuna_remake['descripcion_nse'] = None
```

```python
a = df_comuna_remake.copy()
```

```python
a.info(verbose=True)
```

```python
#Clase alta
a.loc[(a['hogares_calidad_de_construccion'] <= 60) & ((a['hogares_jefes_universitario'] > 40) & (a['hogares_personal_con_cama'] > 3)), 'descripcion_nse'] = 'Alta'  
#Clase media alta
a.loc[((a['hogares_calidad_de_construccion'] <= 60) & ((a['hogares_jefes_universitario'] <= 40) | (a['hogares_personal_con_cama'] <= 3)) & ((a['hogares_piso_precario'] < 10) & (a['personas_utilizan_compu'] > 75)) & ((a['hogares_servicios'] < 10) & (a['hogares_jefe_con_secundaria'] > 60))), 'descripcion_nse'] = 'Media Alta'  
#clase media
a.loc[((a['hogares_calidad_de_construccion'] <= 60) & ((a['hogares_jefes_universitario'] <= 40) | (a['hogares_personal_con_cama'] <= 3)) & ((a['hogares_piso_precario'] < 10) & (a['personas_utilizan_compu'] > 75)) & ((a['hogares_servicios'] >= 10) | (a['hogares_jefe_con_secundaria'] <= 60))), 'descripcion_nse'] = 'Media'  
#clase media baja
a.loc[((a['hogares_calidad_de_construccion'] <= 60) & ((a['hogares_jefes_universitario'] <= 40) | (a['hogares_personal_con_cama'] <= 3)) & ((a['hogares_piso_precario'] >= 10) | (a['personas_utilizan_compu'] <= 75)) & ((a['hogares_piso_precario'] <= 90) & (a['personas_utilizan_compu'] > 75))), 'descripcion_nse'] = 'Media Baja'  
#clase baja integrada
a.loc[((a['hogares_calidad_de_construccion'] <= 60) & ((a['hogares_jefes_universitario'] <= 40) | (a['hogares_personal_con_cama'] <= 3)) & ((a['hogares_piso_precario'] >= 10) | (a['personas_utilizan_compu'] <= 75)) & ((a['hogares_piso_precario'] > 90) | (a['personas_utilizan_compu'] <= 75))), 'descripcion_nse'] = 'Media Baja Integrada'  
#clase baja
a.loc[((a['hogares_calidad_de_construccion'] > 60) & (a['personas_jovenes_asistencia_sin_secundario'] >= 60)), 'descripcion_nse'] = 'Baja'  
#clase muy baja
a.loc[((a['hogares_calidad_de_construccion'] > 60) & (a['personas_jovenes_asistencia_sin_secundario'] < 60)), 'descripcion_nse'] = 'Muy Baja'  
```

```python
df_NSE_final = a.copy()
```

```python
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Alta', 'NSE'] = 1
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Media Alta', 'NSE'] = 2
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Media', 'NSE'] = 3
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Media Baja', 'NSE'] = 4
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Media Baja Integrada', 'NSE'] = 5
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Baja', 'NSE'] = 6
df_NSE_final.loc[df_NSE_final['descripcion_nse'] == 'Muy Baja', 'NSE'] = 7
```

```python
df_NSE_final['NSE'] = df_NSE_final['NSE'].astype(int)
df_NSE_final
```
